```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
```

```python
path = '../../RBNdata/N10K3S100G10001p0.5eps0.1alp2/'
name = 'S????????B????????I????????'

file_names = np.sort(glob.glob(path + '/AFS/' + name))
print(len(file_names))
# Finding Attractor length
AL0 = []
AL1 = []
AL10 = []
AL100 = []
AL1000 = []
AL10000= []
for file_name in file_names:
    l = 0
    with open(file_name, 'r') as file:
        for line in file:
            
            if not line.startswith('G'):
                l = l+1
                for element in line.split():
                    #Split the line into individual integers and convert them to integers
                    if l == 1:
                        AL0.append(int(element))
                    elif l == 2:
                        AL1.append(int(element))
                    elif l == 3:
                        AL10.append(int(element))
                    elif l == 4:
                        AL100.append(int(element))
                    elif l == 5:
                        AL1000.append(int(element))
                    elif l == 6:
                        AL10000.append(int(element))
AL0 = np.array(AL0)
AL1= np.array(AL1)
AL10 = np.array(AL10)
AL100= np.array(AL100)
AL1000 = np.array(AL1000)
AL10000= np.array(AL10000)
```

```python
print(len(AL0), len(AL1), len(AL10), len(AL100), len(AL1000), len(AL10000))
histAL0, binsAL0 = np.histogram(AL0, range=(1,101), bins=100)
histAL1, binsAL1 = np.histogram(AL1, range=(1,101), bins=100)
histAL10, binsAL10 = np.histogram(AL10, range=(1,101), bins=100)
histAL100, binsAL100 = np.histogram(AL100, range=(1,101), bins=100)
histAL1000, binsAL1000 = np.histogram(AL1000, range=(1,101), bins=100)
histAL10000, binsAL10000 = np.histogram(AL10000, range=(1,101), bins=100)

plt.figure(figsize=(10,8))
plt.loglog(range(1,101), histAL0, label="G 0")
plt.loglog(range(1,101), histAL1, label="G 1")
plt.loglog(range(1,101), histAL10, label="G 10")
plt.loglog(range(1,101), histAL100, label="G 100")
plt.loglog(range(1,101), histAL1000, label="G 1000")
plt.loglog(range(1,101), histAL10000, label="G 10000")
plt.legend()
plt.xlabel('AL(Attractor length)')
plt.ylabel('Num(AL) (Total number of AL)')
plt.title('AL vs. Num(AL)\nQS=%d\nN10 K3 S1000 p0.5 eps0.1 alp2' %(len(file_names)))
plt.show()

plt.figure(figsize=(10,8))
plt.semilogy(range(1,101), histAL0, label="G 0")
plt.semilogy(range(1,101), histAL1, label="G 1")
plt.semilogy(range(1,101), histAL10, label="G 10")
plt.semilogy(range(1,101), histAL100, label="G 100")
plt.semilogy(range(1,101), histAL1000, label="G 1000")
plt.semilogy(range(1,101), histAL10000, label="G 10000")
plt.legend()
plt.xlabel('AL(Attractor length)')
plt.ylabel('Num(AL) (Total number of AL)')
plt.title('AL vs. Num(AL)\nQS=%d\nN10 K3 S1000 p0.5 eps0.1 alp2' %(len(file_names)))
plt.show()

plt.figure(figsize=(10,8))
plt.plot(range(1,101), histAL0, label="G 0")
plt.plot(range(1,101), histAL1, label="G 1")
plt.plot(range(1,101), histAL10, label="G 10")
plt.plot(range(1,101), histAL100, label="G 100")
plt.plot(range(1,101), histAL1000, label="G 1000")
plt.plot(range(1,101), histAL10000, label="G 10000")
plt.legend()
plt.xlabel('AL(Attractor length)')
plt.ylabel('Num(AL) (Total number of AL)')
plt.title('AL vs. Num(AL)\nQS=%d\nN10 K3 S1000 p0.5 eps0.1 alp2' %(len(file_names)))
plt.show()
```
