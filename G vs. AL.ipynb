```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
path = '../../RBNdata/N10K3S1000G10001p0.5eps0.1alp2/'
name = 'S????????B????????I????????'

file_names = np.sort(glob.glob(path + '/AFS/' + name))
print(len(file_names))

data = []
# Initialize an empty array to store the lines
avg_att_g = 0
for file_name in file_names:
    data = []
    # Open the file and read line by line
    with open(file_name, 'r') as file:
        for line in file:
        # Check if the line starts with "Generation"
            if not line.startswith('G'):
                # Split the line into individual integers and convert them to integers
                elements = [int(element) for element in line.split()]
                
                # averageing Attractor length for same Generation
                data.append(np.average(elements))
    data = np.array(data)
    avg_att_g = avg_att_g + data
    
# averageing Attractor length for different relization
avg_att_g = avg_att_g/len(file_names)
plt.figure(figsize=(10,6))
plt.title("G vs. <L> \n N10 K3 p0.5 S1000 G10000 eps0.1 alp2\nDifferent structure, boolean sets & conditions")
plt.xlabel("G (Generation)")
plt.ylabel("<L> (Avg Attractor L for 1000 populations)" )
plt.plot(data, label="one ensemble")
plt.plot(avg_att_g, label="%d ensembles" %(len(file_names)))
plt.legend()
plt.show()
```
