```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
path = '../../RBNdata/N10K3S1000G10001p0.5eps0.1alp-2/'
name = 'S????????B????????I????????'

file_names = np.sort(glob.glob(path + '/MFS/' + name))
print(len(file_names))

data = []
# Initialize an empty array to store the lines
avg_MI_g = 0
for file_name in file_names:
    data = []
    # Open the file and read line by line
    with open(file_name, 'r') as file:
        for line in file:
        # Check if the line starts with "Generation"
            if not line.startswith('G'):
                # Split the line into individual integers and convert them to integers
                elements = [float(element) for element in line.split()]
                data.append(np.average(elements))
    data = np.array(data)
    avg_MI_g = avg_MI_g + data
avg_MI_g = avg_MI_g/len(file_names)

plt.figure(figsize=(10,6))
plt.title("G vs. <MI>\n N10 K3 p0.5 S1000 G10000 eps0.1 alp-2\nDifferent structure, boolean sets & conditions")
plt.xlabel("G (Generation)")
plt.ylabel("<MI> (Avg MI for 1000 populations)")
plt.plot(data, label="one ensemble")
plt.plot(avg_MI_g, label="%d ensembles" %(len(file_names)))
plt.legend()
plt.show()
```
